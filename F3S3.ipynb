```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "white")
```

```python
df = pd.read_csv("/content/drive/MyDrive/Bedu/Fase3/Datasets/athlete_olympic_events-clean.csv",index_col=0)
```

```python
df.head()
```

```python
ax = sns.distplot(df['age'],kde=True, norm_hist=False)
ax.set(title="Edades de los Atletas que participaron en los juego Olimpicos",xlabel="Edad",ylabel="Conteo")
```

```python
ax = sns.displot(df['age'],kde=False)
ax.set(title="Edades de los Atletas que participaron en los juego Olimpicos",xlabel="Edad",ylabel="Conteo");
```

```python
por_atleta = df.groupby(level=0)[['age','height','weight']].mean()
por_genero = df.groupby(level=0)['sex'].last()
unidos = por_atleta.merge(por_genero,left_index=True,right_index=True)
hombres=por_atleta[unidos['sex']=='M']
mujeres=por_atleta[unidos['sex']=='F']
mujeres.head()
```

```python
sns.kdeplot(mujeres['height'],shade=True, label="Mujeres")
sns.kdeplot(hombres['height'],shade=True, label="Hombres")
plt.title("Distribuciones de altura por gnero")
plt.xlabel("Altura")
plt.ylabel("Cantidad")
plt.legend(loc="upper right");
```

```python
fig = plt.figure(figsize=(15,15))
sns.kdeplot(mujeres['height'],shade=True, label="Mujeres")
sns.kdeplot(hombres['height'],shade=True, label="Hombres")
plt.title("Distribuciones de altura por gnero")
plt.xlabel("Altura")
plt.ylabel("Cantidad")
plt.legend(loc="upper right");
```

```python
sns.kdeplot(data=unidos, x="height",hue="sex",common_norm=False,shade=True)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020-Santander/main/Datasets/diabetes-clean.csv', index_col=0)
df.head()
```

```python
g1 = sns.distplot(df['glucose'], kde=False, norm_hist=False)
g1.set(title="Glucosa en pacientes con Diabetes",xlabel="Glucosa Medida",ylabel="Conteo");
```

```python
pregnancy_counts = df['pregnancies'].value_counts()
g2 = sns.barplot(pregnancy_counts.index, pregnancy_counts)
g2.set(title="Embarazos",xlabel="No. Embarazos",ylabel="Cantidad de personas");
```

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True},label="Salusable")
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True},label="Enfermas")

plt.title("Densidad de Glucosa en Pacientes con Diabetes")
plt.xlabel("Cantidad de Glucosa")
plt.ylabel("Densidad de Glucosa")
plt.legend(loc="upper right");
```

```python
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True},label="Presin")
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True},label="Glucosa")
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True},label="Insulina")

plt.title("Personas Enfermaz")
plt.xlabel("Mediciones")
plt.ylabel("Densidad")
plt.legend(loc="upper right");
```

```python
df = pd.read_json("https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020-Santander/main/Datasets/zomato_reviews-clean.json")
df.head()
```

```python
df['user_rating'].unique()
```

```python
df['user_rating'].nunique()
```

```python
counts = df['user_rating'].value_counts()
```

```python
ax = sns.barplot(x=counts.index, y=counts)
ax.set_title("Conteo de Rating de Restaurantes en Zomato")
```

```python
porcentajes = counts * 100 / counts.sum()
```

```python
ax = sns.barplot(x=porcentajes.index, y=porcentajes)
ax.set_title("porcentajes de ratings en restaurantes")
ax.set(ylabel="Porcentaje")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 50);
```

```python
#Horizontal
ax = sns.barplot(y=porcentajes.index, x=porcentajes,orient="h")
ax.set_title("porcentajes de ratings en restaurantes")
ax.set(ylabel="Porcentaje");
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,10),sharey=False)
sns.barplot(x=crosstab.columns,y = crosstab.loc[1],ax=axes[0,0])
sns.barplot(x=crosstab.columns,y = crosstab.loc[2],ax=axes[0,1])
sns.barplot(x=crosstab.columns,y = crosstab.loc[3],ax=axes[1,0])
sns.barplot(x=crosstab.columns,y = crosstab.loc[4],ax=axes[1,1])
```
